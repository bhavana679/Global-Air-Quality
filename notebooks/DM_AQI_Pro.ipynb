# How do pollution levels differ between industrial and residential cities, and what patterns characterize global air quality throughout 2024?

```python
import pandas as pd
import numpy as np

!gdown "https://drive.google.com/uc?id=129QmbtoDFhjc21HoHVIPutKuxcbOBOsE" -O City_Types.csv
inres=pd.read_csv("City_Types.csv")
dfinres=(inres)
dfWorldcopy=(dfinres.copy())
```

```python
dfinres.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

dfinres['Date'] = pd.to_datetime(dfinres['Date'])
dfinres['Month'] = dfinres['Date'].dt.month

pollutant_columns = ['CO', 'NO2', 'SO2', 'O3', 'PM2.5', 'PM10']
for col in pollutant_columns:
    dfinres[col] = pd.to_numeric(dfinres[col], errors='coerce')

monthly_avg_pollution_by_type = dfinres.groupby(['Type', 'Month'])[pollutant_columns].mean().reset_index()

#  CALCULATE OVERALL AVERAGES
industrial_overall_avg = monthly_avg_pollution_by_type[
    monthly_avg_pollution_by_type['Type'] == 'Industrial'
]['PM2.5'].mean()

residential_overall_avg = monthly_avg_pollution_by_type[
    monthly_avg_pollution_by_type['Type'] == 'Residential'
]['PM2.5'].mean()

# CREATE COMPARISON DATAFRAME
pm25_comparison = pd.DataFrame({
    'City Type': ['Industrial', 'Residential'],
    'Average PM2.5': [industrial_overall_avg, residential_overall_avg]
})

plt.figure(figsize=(8, 6))
plt.bar(pm25_comparison['City Type'], pm25_comparison['Average PM2.5'])
plt.title('Overall Average PM2.5 Levels in Industrial vs. Residential Cities')
plt.ylabel('Average PM2.5 Level')
plt.xlabel('City Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Thus the industrial cities have higher PM2.5 than residential cities.

```python
import matplotlib.pyplot as plt
import seaborn as sns
monthly_pm25_diff = monthly_avg_pollution_by_type[monthly_avg_pollution_by_type['Type'] == 'Industrial'][['Month', 'PM2.5']].set_index('Month') \
                    - monthly_avg_pollution_by_type[monthly_avg_pollution_by_type['Type'] == 'Residential'][['Month', 'PM2.5']].set_index('Month')

monthly_pm25_diff = monthly_pm25_diff.reset_index()
monthly_pm25_diff.rename(columns={'PM2.5': 'PM2.5_Difference'}, inplace=True)

plt.figure(figsize=(10, 6))
sns.lineplot(
    data=monthly_pm25_diff,
    x='Month',
    y='PM2.5_Difference',
    marker='o',
    color='purple'
)
plt.title('Monthly Difference in PM2.5 (Industrial - Residential)')
plt.ylabel('PM2.5 Difference')
plt.xlabel('Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.axhline(0, color='red', linestyle='--', linewidth=0.8)
plt.show()
```

**This shows that higher the value, the more polluted the industrial areas are compared to residential areas for that month.**

***Monthly Interpretation***

**January  Very High Difference (~85)**

Industrial areas have much higher PM2.5 levels compared to residential.
Reason: winter, heating emissions, stagnant air conditions.

**FebruaryApril  Sharp Decline**

Difference drops to ~30 by April.
Pollution gap reduces  possibly due to:
*   Warming weather
*   Reduced heating
*   Better atmospheric dispersion

**MayAugust  Stable low difference**

Stays between 3338.

Industrial + residential pollution are relatively stable.

**SeptemberDecember  Increase Again**

**SeptemberOctober  gradual rise**

**November (~78)  second peak**

**December (~73)  still high**

**This is typical: winter pollution increases again  industrial emissions + bad dispersion.**

```python
# Calculating daily average pollutant levels by city classification
# Calculating monthly average pollutant levels by city classification

pollutant_columns = ['CO', 'NO2', 'SO2', 'O3', 'PM2.5', 'PM10']
daily_avg_pollution_by_type = dfinres.groupby(['Type', 'Date'])[pollutant_columns].mean().reset_index()
monthly_avg_pollution_by_type = dfinres.groupby(['Type', 'Month'])[pollutant_columns].mean().reset_index()

print("Monthly Average Pollutant Levels by City Type (first 5 rows):")
print(monthly_avg_pollution_by_type.head())
```

```python
# Plotting monthly average pollution levels for each city type and pollutant

import matplotlib.pyplot as plt
import seaborn as sns
pollutant_columns = ['CO', 'NO2', 'SO2', 'O3', 'PM2.5', 'PM10']

fig, axes = plt.subplots(nrows=len(pollutant_columns), ncols=1, figsize=(12, 5 * len(pollutant_columns)), sharex=True)
fig.suptitle('Monthly Average Pollutant Levels in Industrial vs. Residential Cities', fontsize=16, y=0.99)

for i, pollutant in enumerate(pollutant_columns):
    sns.lineplot(
        data=monthly_avg_pollution_by_type,
        x='Month',
        y=pollutant,
        hue='Type',
        marker='o',
        ax=axes[i]
    )
    axes[i].set_title(f'{pollutant} Monthly Average')
    axes[i].set_ylabel(f'{pollutant} Level')
    axes[i].set_xticks(range(1, 13))
    axes[i].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    axes[i].legend(title='City Type')
    axes[i].grid(True)

plt.xlabel('Month')
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

### Across almost all pollutants (CO, NO, SO, PM2.5, PM10):

Industrial Cities consistently show higher average levels for most pollutants indicating a stronger impact of industrial activities and potentially higher population density/traffic.

While Residential Cities follow similar seasonal trends, their absolute pollutant levels are generally lower, suggesting that industrial emissions significantly amplify pollution compared to purely residential sources.

This indicates:
*   more vehicular exhaust
*   more factory emissions
*   higher combustion-related activities

Residential cities stay relatively stable and far below industrial levels.

### Pollutants like CO, NO, PM2.5, PM10 show:

Higher concentrations in winter months (NovFeb)

A clear dip in summer/monsoon months (JunAug)

### This suggests:

Winter inversion effects trap pollutants closer to the ground

Summer winds and rain disperse pollutants effectively

### Pollutants spiking in Winter (Jan, Feb, Nov, Dec): CO, NO2, SO2, PM2.5, PM10.
These are generally associated with combustion processes (heating, industrial activity, traffic) and atmospheric conditions that limit pollutant dispersion (e.g., inversions common in winter).

### Pollutants spiking in Summer (Jun, Jul, Aug): O3.
Ozone formation is highly dependent on sunlight and warm temperatures, which are characteristic of summer months.

```python
pollutant_columns = ['CO', 'NO2', 'SO2', 'O3', 'PM2.5', 'PM10']
industrial_overall_avg = monthly_avg_pollution_by_type[monthly_avg_pollution_by_type['Type'] == 'Industrial'][pollutant_columns].mean()
residential_overall_avg = monthly_avg_pollution_by_type[monthly_avg_pollution_by_type['Type'] == 'Residential'][pollutant_columns].mean()
differences = (industrial_overall_avg - residential_overall_avg).abs()
largest_difference_pollutant = differences.idxmax()
max_difference_value = differences.max()

print("Overall Average Pollutant Levels for Industrial Cities:")
print(industrial_overall_avg)
print("\nOverall Average Pollutant Levels for Residential Cities:")
print(residential_overall_avg)
print(f"\nPollutant with the largest average difference: {largest_difference_pollutant} (Difference: {max_difference_value:.2f})")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=differences.index, y=differences.values, palette='viridis')
plt.title('Absolute Average Difference in Pollutant Levels (Industrial vs. Residential)')
plt.xlabel('Pollutant')
plt.ylabel('Absolute Average Difference')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
